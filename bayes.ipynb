Positive Correlations Groups

```python
from collections import defaultdict
def positive_correlations_group(df):
    cor = df.corr()
    positive_correlations = defaultdict(lambda: [])
    for c in df.columns:
        series = cor[[c]].drop(c)
        series = series.dropna()
        series = series.transpose()
        var_corr_group = [ col for col in series.columns if series[col].item() >= 0.8]
        positive_correlations[c] = var_corr_group
    return positive_correlations
```

Negative Correlation Groups

```python
from collections import defaultdict
def negative_correlations_group(df):
    cor = df.corr()
    negative_correlations = defaultdict(lambda: [])
    for c in df.columns:
        series = cor[[c]].drop(c)
        series = series.dropna()
        series = series.transpose()
        var_corr_group = [ col for col in series.columns if series[col].item() <= -0.8]
        negative_correlations[c] = var_corr_group
    return negative_correlations
```

Now we will make the groups of the variables that have a strong correlation in the same direction.
So first we will do grups for positive correations and then for negative ones.

```python
def purgue(el, correlations):
    dd = defaultdict(lambda x : [])
    for e in el:
        dd[e] = correlations[e]
    return dd
```

We make an algorithm to constract Bayesian Nets with random variables
and get the best model of all the iterations.


```python
import random
from pomegranate import *
import numpy
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import networkx as nx

def get_rnd_cols(df, num_vars = 10):
    df_cols = list(df.columns)
    columns = []
    i = 0
    while i < num_vars:
        num = random.randint(0,len(df_cols)-1)
        if df_cols[num] not in columns:
            columns.append(df_cols[num])
            i += 1
    return df[columns]
    

def random_bayes(training_x, training_y):
    training = training_x.join(training_y)
    model = BayesianNetwork.from_samples(training, algorithm='exact')
    return model

def construct_test_set(X_test):
    to_predict = []
    for test in X_test.values.tolist():
        test.append(None)
        to_predict.append(test)
    return to_predict

def get_prediction_result(results, expected, value):
    score = 0;
    for i,result in enumerate(results):
        if result[-1] == value and result[-1] == expected[i]:
            score +=1
    number_of_values = len([x for x in expected if x == value])
    return score / number_of_values

def get_general_prediction_result(results, expected):
    score = 0;
    for i,result in enumerate(results):
        if result[-1] == expected[i]:
            score +=1
    number_of_values = len(expected)
    return score / number_of_values

def delete_half_negative(df, negative_cases):
    half_negative = negative_cases / 2
    for i in range(half_negative):
        df.sample()
    
def draw_bayesian_network(graph):
    G = nx.DiGraph()
    for i,node in enumerate(graph):
        G.add_node(i)
    for i,node in enumerate(graph):
        for edge in node:
            G.add_edge(edge,i)
    nx.draw(G, with_labels=True, arrowsize=10)
    plt.show()

best_result = 0
best_pos = 0
best_ns = 0
best_neg=0
df = df.dropna(subset=['final_diagnosis_code'])
y = df['final_diagnosis_code']
df = df.drop('final_diagnosis_code', axis = 1)
for i in range(1000):
    X = get_rnd_cols(df)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
    X_test.columns
    model = random_bayes(X_train, y_train)
    to_predict = construct_test_set(X_test)
    to_predict
    try:
        results = model.predict(to_predict)
    except ValueError:
        continue
    if get_general_prediction_result(results, y_test.values.tolist()) >= best_result:
        best_columns = X.columns
        best_result = get_general_prediction_result(results, y_test.values.tolist())
        best_pos = get_prediction_result(results, y_test.values.tolist(),1)
        best_ns = get_prediction_result(results, y_test.values.tolist(),2)
        best_neg = get_prediction_result(results, y_test.values.tolist(),3)
        best_model = model
print(best_columns)
draw_bayesian_network(best_model.structure)
print("General prediction result")
print(best_result)
print("Positive prediction result")
print(best_pos)
print("Don't know prediction result")
print(best_ns)
print("Negative prediction result")
print(best_neg)
```

But also it has limitations, and that is mainly because the algorithm that constructs the bayesian net is of exponential cost, making the number of variables that you can work with limitated. In fact, in the docs of the library that we used, advises us that we should not try to construct a net of more than 25 variables.
